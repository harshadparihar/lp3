```python
def remove_outlier(df1,col):
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3-Q1
    lower_whisker = Q1-1.5*IQR
    upper_whisker = Q3+1.5*IQR
    df[col] = np.clip(df1[col],lower_whisker,upper_whisker)
    return df1
def trear_outliers_all(df1,col_list):
    for c in col_list:
        df1 = remove_outlier(df,c)
    return df1
```

```python
df = trear_outliers_all(df,df.iloc[:,0::])
df.plot(kind = "box",subplots = True,layout = (7,2),figsize = (15,20))
```

```python
import haversine as hs
travel_dist = []
for pos in range(len(df['pickup_longitude'])):
    long1,lati1,long2,lati2 = [df['pickup_longitude'][pos],df['pickup_latitude'][pos],df['dropoff_longitude'][pos],df['dropoff_latitude'][pos]]
    loc1=(lati1,long1)
    loc2=(lati2,long2)
    c = hs.haversine(loc1,loc2)
    travel_dist.append(c)
print(travel_dist)
df['dist_travel_km'] = travel_dist
df.head()
    
```

```python
df = df.loc[(df.dist_travel_km >=1) | (df.dist_travel_km<=130)]
print("Remaining observations in the dataset:", df.shape)
```

```python
incorrect_coordinates = df.loc[(df.pickup_latitude > 90) | (df.pickup_latitude < -90) | (df.dropoff_latitude > 90) | (df.dropoff_latitude < -90) | (df.pickup_longitude > 180) | (df.pickup_longitude < -180) |(df.dropoff_longitude > 90) |(df.dropoff_longitude < -90)]
```

```python
x = df[['pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count','hour','day','month','year','dayofweek','dist_travel_km']]
```

```python
y = df['fare_amount']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.33)
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
regression.intercept_
```

```python
regression.coef_
```

```python
prediction = regression.predict(X_test)
```

```python
print(prediction)
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error , mean_absolute_error
R2 = r2_score(y_test,prediction)
MSE = mean_squared_error(y_test,prediction)
MAE = mean_absolute_error(y_test,prediction)
RMSE = np.sqrt(MSE)
print("r2-score:",R2,"\nMSE: ",MSE,"\nMAE: ",MAE,"\nRMSE: ",RMSE)
```
